## **TEP VS DeepSentiPers (IT domain)**

```
!pip3 install torch torchvision
```

```
!pip install -U sentence-transformers
```

```
from sentence_transformers import SentenceTransformer, util
name = 'https://public.ukp.informatik.tu-darmstadt.de/reimers/sentence-transformers/v0.2/xlm-r-100langs-bert-base-nli-mean-tokens.zip'
```

```
model = SentenceTransformer(model_name_or_path = name, device = 'cuda')
```

```
print("Max Sequence Length: ", model.max_seq_length)
```

```
import pickle
#Load sentences & embeddings from disc
with open('/content/drive/MyDrive/Multilingual_Multimodal_Universal_MT/Domain_Adaptation/SentenceBERT/embeddings.pkl', "rb") as fIn:
    stored_data = pickle.load(fIn)
    parallel_corpus_sentences = stored_data['sentences']
    parallel_corpus_embeddings = stored_data['embeddings']
```

```
parallel_corpus_embeddings = parallel_corpus_embeddings[:5561]
```

```
print(type(parallel_corpus_embeddings))
print(parallel_corpus_embeddings.shape)
```

```
print(type(DeepSentiPers_sentences))
print(len(DeepSentiPers_sentences))
```

```
DeepSentiPers_embeddings = model.encode(DeepSentiPers_sentences, show_progress_bar=True, convert_to_tensor=True)
```

```
import torch
parallel_corpus_embeddings = torch.from_numpy(parallel_corpus_embeddings)
```

```
print(type(DeepSentiPers_embeddings))
print(type(parallel_corpus_embeddings))
```

```
#Compute cosine-similarits
cosine_scores = util.pytorch_cos_sim(parallel_corpus_embeddings, DeepSentiPers_embeddings)
```

```
#Output the pairs with their score
for i in range(len(parallel_corpus_embeddings)):
  if cosine_scores[i][i] > 0.85:
    print("{} \t\t {} \t\t Score: {:.4f}".format(parallel_corpus_sentences[i], DeepSentiPers_sentences[i], cosine_scores[i][i]))
```

```
queries = DeepSentiPers_sentences
```

```
data = [] # Just to save the result
# Find the closest 5 sentences of the corpus for each query sentence based on cosine similarity
top_k = min(5, len(parallel_corpus_sentences))
for query in queries:
    query_embedding = model.encode(query, convert_to_tensor=True)

    # We use cosine-similarity and torch.topk to find the highest 5 scores
    cos_scores = util.pytorch_cos_sim(query_embedding, parallel_corpus_embeddings)[0]
    top_results = torch.topk(cos_scores, k=top_k)

    print("\n\n======================\n\n")
    data.append("\n\n======================\n\n")
    print("Query:", query)
    data.append("Query:" + str(query))
    print("\nTop 5 most similar sentences in corpus:")
    data.append("\nTop 5 most similar sentences in corpus:")
    
    for score, idx in zip(top_results[0], top_results[1]):
        print(parallel_corpus_sentences[idx], "(Score: {:.4f})".format(score))
        data.append(str(parallel_corpus_sentences[idx]) + "(Score: {:.4f})".format(score))
```

```
# Number of sentences in DeepSentiPers 
print(len(data)/8)
```

```
df = pd.DataFrame(data)
```

```
# Save df to csv
df.to_csv("/content/drive/MyDrive/Multilingual_Multimodal_Universal_MT/Domain_Adaptation/DataAdaptation/DeepSentiPers-TEP.csv")
```
